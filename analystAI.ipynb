```python
bags = []
header_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
HEADERS=({'User-Agent':header_agent,'Accept-Language':'en-US, en;q=0.5'})

for i in range(1,21): 
    
     url = f"https://www.amazon.in/s?k=bags&page={i}&crid=2M096C61O4MLT&qid=1690033417&sprefix=ba%2Caps%2C283&ref=sr_pg_{i}.html"
     response = requests.get(url,headers=HEADERS)
     soup = BeautifulSoup(response.content,'html.parser')
     div = soup.find_all('div',attrs={'class':'sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 sg-col-12-of-24 s-list-col-right'})
    
     for d in div:
         ds = d.find('div',class_='a-section a-spacing-small a-spacing-top-small');
         try:
             link = d.find('a',class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal').get('href')
             product_url = "https://amazon.in"+link 
             name = d.find('span',class_='a-size-medium a-color-base a-text-normal').text.strip() 
             price = d.find('span',{'class':'a-price-whole'}).text.strip()
             rating = d.find('span',class_='a-icon-alt').text.strip()
             rating = rating[0:3]
             review = d.find('span',class_='a-size-base s-underline-text').text.strip()
             bags.append([product_url,name,price,rating,review])
         except:
            print(".")
            continue
         
len(bags)       
```

```python
df = pa.DataFrame(bags,columns=['Product Url','Product Name','Product Price','Ratings','Reviews'])
```

```python
links = [];
header_agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
HEADERS=({'User-Agent':header_agent,'Accept-Language':'en-US, en;q=0.5'})


for i in range(1,21):

    url = f"https://www.amazon.in/s?k=bags&page={2}&crid=2M096C61O4MLT&qid=1690033417&sprefix=ba%2Caps%2C283&ref=sr_pg_{2}.html"
    response = requests.get(url,headers=HEADERS)
    soup = BeautifulSoup(response.content,'html.parser')
    div = soup.find_all('div',attrs={'class':'sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20 sg-col-12-of-24 s-list-col-right'})
            
    for d in div:
            ds = d.find('div',class_='a-section a-spacing-small a-spacing-top-small');
            try:
                link = d.find('a',class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal').get('href');
                product_url = "https://amazon.in"+link
                links.append(product_url)
            except:
                continue
                
len(links)           
```

```python
info = [];

for l in links:
    
    try:
        result = requests.get(l,headers=HEADERS)
        soup = BeautifulSoup(result.content,'html.parser')

        # title
        title = soup.select_one("#productTitle").text.strip()

        # price
        price = soup.find('span',{'class':'a-price aok-align-center reinventPricePriceToPayMargin priceToPay'})
        price = price.find('span',{'class':'a-price-whole'}).text.strip()

        # rating
        rating = soup.select_one("#acrPopover > span > a > span").text.strip()

        # review
        review = soup.select_one("#acrCustomerReviewText").text.strip()
        x = np.array(review.split())
        res = x[np.char.isnumeric(x)].astype(int)
        res = res[0]
        
        # Description
        feature = soup.select("#feature-bullets > ul > li")
        full=[]
        for f in feature:
            detail = f.select_one("span").text
            full.append(detail)
            desc = ''.join(full).strip()
        
        # ASIN number
        ul = soup.find('ul',{'class':'a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list'})
        li = ul.find_all('li')
        asin = li[3].select_one("span > span:last-child").text.strip()

        # Manufacturur
        manu = li[2].select_one("span > span:last-child").text.strip()
            
        # Product description
        prod_desc = soup.select_one("#productDescription > p > span").text.strip()

        info.append([product_url,title,price,rating,res,desc,asin,prod_desc,manu])
    
    except:
        continue  
```

```python
df2 = pa.DataFrame(info,columns=['Link','Title','Price','Rating','Reviews','Description','ASIN','Product_desc','Manufacturer'])
```
